```python
# enter_raw = pd.read_csv('../data/marking/enter-test.csv', parse_dates=['dt'])
# enter_int = enter_raw.rename(columns={ 'cnt': 'enter_cnt' }).set_index('dt')
# enter_agg = enter_int.groupby('dt').sum(numeric_only=True)
# enter = enter_agg.resample('D').sum(numeric_only=True)[:-1]
# enter.to_csv('../data/marking/enter-test-aggregate.csv')
enter = pd.read_csv('../data/marking/enter-test-aggregate.csv', parse_dates=['dt'], index_col='dt')
enter_smooth = enter.rolling(7, center=True).mean()
enter_smooth.loc[enter.index].to_csv('../data/marking/enter-test-aggregate-smooth.csv')
enter_std = enter_smooth['enter_cnt'].std()
enter_norm = enter_smooth.copy()
enter_norm['enter_cnt'] = enter_norm['enter_cnt'] / enter_std
```

```python
tmp1 = pd.read_csv('../data/marking/enter-aggregate.csv', parse_dates=['dt'], index_col='dt')
tmp2 = pd.read_csv('../data/marking/enter-test-aggregate.csv', parse_dates=['dt'], index_col='dt')
tmp3 = pd.concat([tmp1, tmp2])
tmp4 = tmp3.rolling(7, center=True).mean()
tmp4.to_csv('../data/marking/enter-full-aggregate-smooth.csv')
px.line(tmp4)
```

```python
tmp = enter.copy()
tmp['enter_cnt_smooth'] = enter_smooth['enter_cnt']
px.line(tmp)
```
