```python
import os
import glob

path = os.getcwd()

yandex_json_files = glob.glob(os.path.join(path, "yandex-wordstats/*.json"))
google_csv_files = glob.glob(os.path.join(path, "google-trends/*.csv"))

yandex_data = []
google_data = []

start_date = '2021-10-15'
end_date = '2022-11-30'

def prepare_title(f):
    head, tail = f.split('.')
    return os.path.basename(head)

def prepare_data(title, df):
    df = df.drop_duplicates(subset='date')

    df = df.sort_values('date') \
        .loc[df['totalCount'] != 0] \
        .loc[df['date'] >= start_date] \
        .loc[df['date'] <= end_date]

    df['totalCount'] = (df['totalCount']-df['totalCount'].min())/(df['totalCount'].max()-df['totalCount'].min())

    df = df[['date', 'totalCount']]

    df = df.set_index('date')

    return { 'title': title, 'df': df }


for f in yandex_json_files:
    df = pd.read_json(f)
    yandex_data.append(prepare_data(prepare_title(f), df))

for f in google_csv_files:
    df = pd.read_csv(f, parse_dates=['Неделя'])
    df['Баллы'] = df['Баллы'].rolling(window=5, center=True).mean()
    df['date'] = df['Неделя']
    df['totalCount'] = df['Баллы']
    google_data.append(prepare_data(prepare_title(f), df))

yandex_data[5]['df']
```

```python
google_popularity = 0.35
yandex_popularity = 0.65

dfs = []

for data in yandex_data:
    title, yandex_df = data['title'], data['df']

    g_data = [x for x in google_data if x['title'] == title][0]

    google_df = g_data['df']

    yandex_df_spline = yandex_df.resample('D').interpolate('time', order=2)

    google_df_spline = google_df.resample('D').interpolate('time', order=2)

    df = pd.DataFrame({
        'google': google_df_spline['totalCount'],
        'yandex': yandex_df_spline['totalCount'],
        'y0.65/g0.35': yandex_df_spline['totalCount'] * yandex_popularity + google_df_spline['totalCount'] * google_popularity,
    })

    dfs.append({'title': title, 'df': df})
```

```python
for df in dfs[4:]:
    px.line(df['df'], title=df['title']).show()
```
